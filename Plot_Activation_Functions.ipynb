```python
# Sigmoid function
import numpy as np

def sigmoid(z):
    return (1.0/(1.0+np.exp(-z)))

print(sigmoid(10))

print(sigmoid(-10))
```

```python
# Derivative of Sigmoid
import numpy as np

def sigmoid_prime(z):
    s = (1.0/(1.0+np.exp(-z)))
    return (s*(1-s))

print(sigmoid_prime(0))
print(sigmoid_prime(2))
print(sigmoid_prime(10))
print(sigmoid_prime(-2))
print(sigmoid_prime(-10))
```

```python
# Hyperbolic Tangent - tanh function
import numpy as np

def tanh(z):
    return (np.tanh(z))

print(tanh(0))
print(tanh(1))
print(tanh(2))
print(tanh(5))
print(tanh(10))

print(tanh(-1))
print(tanh(-2))
print(tanh(-5))
print(tanh(10))
```

```python
# Differentiated Hyperbolic Tangent - diff_tanh function
import numpy as np

def tanh_prime(z):
    return (1 - (np.tanh(z)**2))

print(tanh_prime(0))
print(tanh_prime(1))
print(tanh_prime(2))
print(tanh_prime(5))
print(tanh_prime(10))

print(tanh_prime(-1))
print(tanh_prime(-2))
print(tanh_prime(-5))
print(tanh_prime(-10))
```

```python
# ReLU - Rectified Linear Unit function
import numpy as np

def reLU(z):
    return (np.maximum(0,z))

print(reLU(0))
print(reLU(1))
print(reLU(2))
print(reLU(5))
print(reLU(10))
```

```python
# Differentiated ReLU - Diff_Rectified Linear Unit function
import numpy as np

def reLU_prime(z):
    if z==0 or z<0:
        z = 0
    else:
        z = 1
    return z

print(reLU_prime(0))
print(reLU_prime(1))
print(reLU_prime(2))
print(reLU_prime(5))
print(reLU_prime(10))

print(reLU_prime(-1))
print(reLU_prime(-2))
print(reLU_prime(-5))
print(reLU_prime(-10))
```

```python
# Leaky ReLU - Leaky Rectified Linear Unit function
import numpy as np

def leakyreLU(z):
    return (np.maximum(0.01*z,z))

print(leakyreLU(0))
print(leakyreLU(1))
print(leakyreLU(2))
print(leakyreLU(5))
print(leakyreLU(10))

print(leakyreLU(-1))
print(leakyreLU(-2))
print(leakyreLU(-5))
print(leakyreLU(-10))
```

```python
# Differentiated Leaky ReLU - Diff. Leaky Rectified Linear Unit function
import numpy as np

def leakyreLU_prime(z):
    if z==0 or z<0:
        z = 0.01
    else:
        z = 1
    return z

print(leakyreLU_prime(0))
print(leakyreLU_prime(1))
print(leakyreLU_prime(2))
print(leakyreLU_prime(5))
print(leakyreLU_prime(10))

print(leakyreLU_prime(-1))
print(leakyreLU_prime(-2))
print(leakyreLU_prime(-5))
print(leakyreLU_prime(-10))
```

```python
# Activation Functions

import numpy as np

def activation_function(z, activation="tanh"):
    if activation == "sigmoid":
        return (1.0/(1.0+np.exp(-z)))
    elif activation == "tanh":
        return (np.tanh(z))
    elif activation == "reLU":
        return (np.maximum(0,z))
    elif activation == "leakyreLU":
        return (np.maximum(0.01*z,z))   
```

```python
print(activation_function(10,"sigmoid"))
print(activation_function(-10,"sigmoid"))

print(activation_function(10,"tanh"))
print(activation_function(-10,"tanh"))

print(activation_function(10,"reLU"))
print(activation_function(-10,"reLU"))

print(activation_function(1,"leakyreLU"))
print(activation_function(2,"leakyreLU"))
print(activation_function(5,"leakyreLU"))
print(activation_function(10,"leakyreLU"))

print(activation_function(0,"leakyreLU"))
print(activation_function(-1,"leakyreLU"))
print(activation_function(-2,"leakyreLU"))
print(activation_function(-5,"leakyreLU"))
print(activation_function(-10,"leakyreLU"))

```

#### ReLU and Its derivative
**ReLU stands for Rectified Linear Unit.**<br />
- f(x) = max(x,0) - the maximum between x and 0 <br />
- Code
**def relu(x) :
    return max(x, 0)**
    
**The derivative of the ReLU is :**<br />
- 1 if x is greater than 0
- 0 if x is smaller or equal to 0
- Code
  **def der_relu(x):**
      if x <= 0 : 
         return 0
      if x > 0:  
         return 1

**Advantage:** 
   - Easy to implement and quick to compute
   
**Disadvantage:** 
   - Problematic when we have lots of negative values, since the outcome is always 0 <br />
   - and leads to the **death of the neuron**

```python
# RelU and Its derivative
import numpy as np
import matplotlib.pyplot as plt

def relu(x) :
    return max(x, 0)

def der_relu(x):
    if x <= 0 :
        return 0
    if x > 0 :
        return 1

# Data which will go through activations
x = np.linspace(-10,10,100)

plt.figure(figsize=(12,8))
plt.plot(x, list(map(lambda x: relu(x),x)), label="relu")
plt.plot(x, list(map(lambda x: der_relu(x),x)), label="derivative")
plt.title("ReLU")
plt.legend()
plt.show()
```

##### Leaky ReLU and Its Derivative
- Leaky-ReLU is an improvement of the main default of the ReLU
- it can handle the negative values pretty well
- f(x) = max(0.01*x, x)
- f(x) = x if x>0 
       = 0.01*x else
       
- Code 
**def leaky_relu(x):**<br />
    return max(0.01*x,x)
    
##### The derivative of Leaky ReLU
- f(x) = 1    if x>0<br /> 
      = 0.01 else
      
- Code  
**def der_leaky_relu(x):**<br>
    if x < 0 :<br />
        return 0.01<br />
    if x >= 0 :<br />
        return 1

Advantage	
- Leaky-ReLU overcomes the problem of the death of the neuron linked to a zero-slope.

Disadvantage	
- The factor 0.01 is arbitraty, and can be tuned (PReLU for parametric ReLU)

```python
# Leaky ReLU and Its Derivative
import numpy as np
import matplotlib.pyplot as plt

def leaky_relu(x):
    return max(0.01*x,x)

def der_leaky_relu(x):
    if x < 0 :
        return 0.01
    if x >= 0 :
        return 1

# Data which will go through activations
x = np.linspace(-10,10,100)

plt.figure(figsize=(12,8))
plt.plot(x, list(map(lambda x: leaky_relu(x),x)), label="leaky-relu")
plt.plot(x, list(map(lambda x: der_leaky_relu(x),x)), label="derivative")
plt.title("Leaky-ReLU")
plt.legend()
plt.show()
```

##### ELU - Exponential Linear Units (ELU)  and It Derivative
- make the mean activations closer to zero, which speeds up learning.
- f(x)
- f'(x)


**Advantage**
- Can achieve higher accuracy than ReLU

**Disadvantage**	
- Same as ReLU, and a needs to be tuned    

```python
# ELU and Its Derivative
# Exponential Linear Units (ELU)

import numpy as np
import matplotlib.pyplot as plt
def elu(x):
    if x > 0 :
        return x
    else : 
        return (np.exp(x)-1)

def der_elu(x):
    if x > 0 :
        return 1
    else :
        return np.exp(x)

# Data which will go through activations
x = np.linspace(-10,10,100)

plt.figure(figsize=(12,8))
plt.plot(x, list(map(lambda x: elu(x),x)), label="elu")
plt.plot(x, list(map(lambda x: der_elu(x),x)), label="derivative")
plt.title("ELU")
plt.legend()
plt.show()
```

##### Softplus and Its Derivative
- The Softplus function is a continuous approximation of ReLU

**Advantage**
- Softplus is continuous and might have good properties in terms of derivability. It is interesting to use it when the values are between 0 and 1.

**Disadvantage**
- As ReLU, problematic when we have lots of negative values, since the outcome gets really close to 0 and might lead to the death of the neuron

```python
# Softplus and Its Derivative

import numpy as np
import matplotlib.pyplot as plt

def softplus(x):
    return np.log(1+np.exp(x))

def der_softplus(x):
    return 1/(1+np.exp(x))*np.exp(x)

# Data which will go through activations
x = np.linspace(-10,10,100)

plt.figure(figsize=(12,8))
plt.plot(x, list(map(lambda x: softplus(x),x)), label="softplus")
plt.plot(x, list(map(lambda x: der_softplus(x),x)), label="derivative")
plt.title("Softplus")
plt.legend()
plt.show()
```

##### Sigmoid

**Advantage**
- Interpretability of the output mapped between 0 and 1, compute gradient quickly

**Disadvantage**
- Vanishing Gradient

```python
# Sigmoid and Its Derivative

import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    return 1/(1+np.exp(-x))

def der_softmax(x):
    return 1/(1+np.exp(-x)) * (1-1/(1+np.exp(-x)))

# Data which will go through activations
x = np.linspace(-10,10,100)

plt.figure(figsize=(12,8))
plt.plot(x, list(map(lambda x: softmax(x),x)), label="softmax")
plt.plot(x, list(map(lambda x: der_softmax(x),x)), label="derivative")
plt.title("Softmax")
plt.legend()
plt.show()
```

##### Hyperbolic Tangent
- Hyperbolic tangent is quite similar to the sigmoid function, expect is maps the input between -1 and 1.

**Advantage**
- Efficient since it has mean 0 in the middle layers between -1 and 1

**Disadvantage**
- Vanishing gradient too


```python
# Hyperbolic Tangent - tanh and Its Derivative

import numpy as np
import matplotlib.pyplot as plt


def hyperb(x):
    return (np.exp(x)-np.exp(-x)) / (np.exp(x) + np.exp(-x))

def der_hyperb(x):
    return 1 - ((np.exp(x)-np.exp(-x)) / (np.exp(x) + np.exp(-x)))**2

# Data which will go through activations
x = np.linspace(-10,10,100)

plt.figure(figsize=(12,8))
plt.plot(x, list(map(lambda x: hyperb(x),x)), label="hyperbolic")
plt.plot(x, list(map(lambda x: der_hyperb(x),x)), label="derivative")
plt.title("Hyperbolic")
plt.legend()
plt.show()
```

##### Arctan

**Advantage**
- Simple derivative

**Disadvantage**
- Vanishing gradient

```python
# Arc Tangent - atan and Its Derivative

import numpy as np
import matplotlib.pyplot as plt

def arctan(x):
    return np.arctan(x)

def der_arctan(x):
    return 1 / (1+x**2)

# Data which will go through activations
x = np.linspace(-10,10,100)

plt.figure(figsize=(12,8))
plt.plot(x, list(map(lambda x: arctan(x),x)), label="arctan")
plt.plot(x, list(map(lambda x: der_arctan(x),x)), label="derivative")
plt.title("Arctan")
plt.legend()
plt.show()
```

```python
def plot(func, yaxis=(-1.4, 1.4)):
    plt.ylim(yaxis)
    plt.locator_params(nbins=5)
    plt.xticks(fontsize = 14)
    plt.yticks(fontsize = 14)
    plt.axhline(lw=1, c='black')
    plt.axvline(lw=1, c='black')
    plt.grid(alpha=0.4, ls='-.')
    plt.box(on=None)
    plt.plot(x, func(x), c='r', lw=3)
```

```python
# Binary Step
binary_step = np.vectorize(lambda x: 1 if x > 0 else 0, otypes=[np.float])
```

```python
plot(binary_step, yaxis=(-0.4, 1.4))
```
